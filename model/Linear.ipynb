```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("freelancer_earnings_bd.csv")

X = df[["Job_Completed", "Hourly_Rate", "Job_Success_Rate", "Client_Rating", "Job_Duration_Days", "Rehire_Rate", "Marketing_Spend"]]
y = df["Earnings_USD"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=537)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_preds = lr_model.predict(X_test)

# mae = mean_absolute_error(y_test, lr_preds)
```

```python
print("Linear Regression R2 Score:", r2_score(y_test, lr_preds))

print("MAE", mean_absolute_error(y_test, lr_preds))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, r2_score

df = pd.read_csv("freelancer_earnings_bd.csv")

X = df[["Job_Completed", "Hourly_Rate", "Job_Success_Rate", "Client_Rating", "Job_Duration_Days", "Rehire_Rate", "Marketing_Spend"]]
y = df["Earnings_USD"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=537)

df["Experience_Level"] = pd.Categorical(
    df["Experience_Level"],
    categories=["Beginner", "Intermediate", "Expert"],
    ordered=True,
)

le = LabelEncoder()
df["Experience_Level_encoded"] = le.fit_transform(df["Experience_Level"])

df.drop(columns=["Experience_Level"], inplace=True)

model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred)}")
```
